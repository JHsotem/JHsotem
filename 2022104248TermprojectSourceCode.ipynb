https://bus.koreacharts.com/city-bus/36020.html 에서 버스노선 및 정류장 정보 추출했습니다.<br>
각 노선 별 웹사이트 주소를 먼저 추출하는 작업입니다. <br>


```python
import xml.etree.ElementTree as ET
import csv
import pandas as pd
import urllib.request
from urllib.parse import quote
import bs4
from bs4 import BeautifulSoup
import requests
import folium 

url = "https://bus.koreacharts.com/city-bus/36020.html"
req = urllib.request.Request(url)
sourcecode = urllib.request.urlopen(url).read()
soup = BeautifulSoup(sourcecode, "html.parser")

urlMatrix = []

for href in soup.find("ul", class_="list-unstyled").find_all("li"):
    urlMatrix.append(href.find("a")["href"])
print(urlMatrix)
```

각 url의 순서별 버스 노선 번호를 list로 만드는 작업입니다. <br>

```python
routeNameMatrix = []

for i in urlMatrix:
    url = 'https://bus.koreacharts.com' + i
    req = urllib.request.Request(url)
    sourcecode = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(sourcecode, "html.parser")

    for h2 in soup.find("div", class_="box box-default").find_all("h2"):
        routeNameMatrix.append(str(h2)[22:-13])
print(routeNameMatrix)
```

각 노선이 정차하는 버스정류장의 데이터를 추출하는 작업입니다. <br>

```python
busStopMatrix = [] 

for i in urlMatrix:
    url = 'https://bus.koreacharts.com' + i
    req = urllib.request.Request(url)
    sourcecode = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(sourcecode, "html.parser")
    temBusStopMatrix = []
    for td in soup.find('section', class_='content').find_all("td"):
        temBusStopMatrix.append(td)
    busStopMatrix.append(temBusStopMatrix)
print(busStopMatrix)
```

busStopMatrix list의 list in list에서 불필요한 문자를 제거하는 작업입니다. <br>

```python
nBusStopMatrix = [] 
for i in busStopMatrix:
    temNBusStopMatrix = []
    temNBusStopMatrix.append(str(i[0]).strip("<td class=\"text-center\">").strip("</td>"))
    for a in range(7, len(i)):
        temNBusStopMatrix.append(str(i[a]).strip("<td class=\"text-center\">").strip("</td>"))
    nBusStopMatrix.append(temNBusStopMatrix[0::2])

print(nBusStopMatrix)
```

여수시 버스노선은 노선 번호만 같고 기점, 종점 및 경유정류장이 다른 버스노선이 존재하므로 이를 정리하는 과정입니다. <br>
위의 routeNameMatrix에서 중복되는 노선만 노선 번호 뒤 기점과 종점을 기입하도록 하겠습니다.<br>

```python
nRouteMatrix = [] 

for e in range(len(routeNameMatrix)):
    if routeNameMatrix.count(routeNameMatrix[e]) == 1:
        nRouteMatrix.append(routeNameMatrix[e])
    else:
        newRouteName = routeNameMatrix[e] + "(" + nBusStopMatrix[e][1] + " - " + nBusStopMatrix[e][-1] + ")"
        nRouteMatrix.append(newRouteName)
print(nRouteMatrix)
```

기점과 종점에 따른 버스노선의 명칭을 모두 정리했으므로 정리된 명칭과 버스가 지나는 정류장 list를 합치는 과정입니다. <br>

```python
fixedBusList = []
for i in range(len(nRouteMatrix)):
    temFixedBusList = []
    temFixedBusList.append(nRouteMatrix[i])
    for a in range(1,len(nBusStopMatrix[i])):
        temFixedBusList.append(nBusStopMatrix[i][a])
    fixedBusList.append(temFixedBusList)
print(fixedBusList)
```

정리한 데이터를 csv 파일로 옮기는 작업입니다.<br>

```python
with open('버스노선_및_정류장_데이터_크롤링.csv', 'w') as newfile:
    myWriter = csv.writer(newfile)
    for i in range(len(fixedBusList)):
        myWriter.writerow(fixedBusList[i][:])
```
