```python
import csv
import pandas as pd

# 여수의 2018년 기준 버스정류장의 명칭과 위치한 지역명이 포함된 데이터를 공공데이터포탈에서 가져왔습니다. 
# 작업환경이 MS WINDOWS이기 떄문에 인코딩을 cp494로 진행했습니다.
df = pd.read_csv('전라남도_여수시_버스정류장정보_20180828..csv', encoding = 'cp949')
df
```

```python
# 데이터에서 필요하지 않은 부분인 연번, 정류장번호, 유형, CCTV, 단말기, 액정, 의자, 광고판, 버스베이를 제거한 새 파일을 만드는 코드입니다.
# 원본 데이터를 가공하여 만든 파일은 다음과 같습니다.

fileMatrix = []
newMatrix = [] 

with open('전라남도_여수시_버스정류장정보_20180828..csv', 'r') as file:
    
    for lineContent in file:
        fileMatrix.append(lineContent.strip('\n').split(','))
    for a in fileMatrix:
        newMatrix.append(a[1:3])

lenNewMatrix = len(newMatrix)

with open('전라남도_여수시_버스정류장정보_20180828_fixed.csv', 'w') as newfile:
    myWriter = csv.writer(newfile)
    for i in range(lenNewMatrix):
        myWriter.writerow(newMatrix[i][:])
        
df = pd.read_csv('전라남도_여수시_버스정류장정보_20180828_fixed.csv', encoding = 'cp949')
df
```

```python
# 지역별로 버스정류장을 분류했습니다. 
# 정류장명칭이 같은 경우 숫자를 붙여 구분지었습니다. 

fileMatrix = []
locationSet = set()

with open('전라남도_여수시_버스정류장정보_20180828_fixed.csv', 'r') as file:
    for lineContent in file:
        fileMatrix.append(lineContent.strip('\n').split(','))

# 위 단계까지 진행했을 때 정류장명칭과 지역명이 들어가있는 list in list 형태에서 사이사이 빈 리스트가 생성되어 텅 빈 리스트를 제거하는 과정입니다. 
# 또한 fileMatrix의 0번째 element인 ['정류장명칭', '지역명']은 필요없으니 제거한다.

while ([''] in fileMatrix):
    fileMatrix.remove([''])
fileMatrix.remove(['정류장명칭', '지역명'])

# 정류장이 위치해있는 지역명을 모두 한 집합에 넣어 확인하고 개수 또한 확인한다. 

for a in fileMatrix:
    b = a[1]
    locationSet.add(b)

print(locationSet)
print(len(locationSet))
```

```python
# 지역별로 몇개의 버스정류장이 있는지 분류하는 작업입니다. 

locationDiction = {}
print(len(fileMatrix))
for a in fileMatrix: 
    b = a[1]
    if (b not in locationDiction):
        locationDiction[b] = 1
    else:
        locationDiction[b] = locationDiction[b] + 1

print(locationDiction)
```
